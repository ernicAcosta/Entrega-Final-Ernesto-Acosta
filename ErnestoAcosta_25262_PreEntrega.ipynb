## ETAPA 1:3. Calidad de datos** (nulos y duplicados)

```python

# ============================================
#  FUNCIN DE CONTROL DE CALIDAD DE DATOS que llamamos "calidad"
# ============================================
# Par谩metros:
#   df: DataFrame de pandas que se analizar谩.
#   nombre: texto descriptivo del DataFrame (ejemplo: 'VENTAS').
#   clave: (opcional) nombre de la columna para buscar duplicados espec铆ficos.

def calidad(df, nombre, clave=None):
    """
    Analiza la calidad del DataFrame:
      - Muestra cantidad de nulos por columna.
      - Cuenta filas duplicadas completas.
      - Si se indica una clave, muestra los valores duplicados m谩s frecuentes.
    Par谩metros:
      df: DataFrame de pandas que se analizar谩.
      nombre: texto descriptivo del DataFrame (ejemplo: 'VENTAS').
      clave: (opcional) nombre de la columna para buscar duplicados espec铆ficos.
    """


    print(f"### {nombre}") #imprime un nombre

    # -------------------------------------------------
    # Mostrar cantidad de valores nulos por columna
    # -------------------------------------------------
    # df.isna() devuelve un DataFrame booleano con True donde hay NaN.
    # .sum() cuenta los True (o sea, los nulos) por columna.
    # .to_frame("nulos") convierte el resultado en un DataFrame con una columna llamada 'nulos'.
    display(df.isna().sum().to_frame("nulos"))

    # -------------------------------------------------
    # Contar filas duplicadas completas
    # -------------------------------------------------
    # df.duplicated(keep=False) marca como True todas las filas que tienen otra igual.
    # keep=False significa que marca todas las copias, no solo una.
    # .sum() cuenta cu谩ntas filas est谩n repetidas.
    dup_rows = df.duplicated(keep=False).sum()
    print("Filas duplicadas (exactas):", dup_rows)

    # -------------------------------------------------
    # Si se especific贸 una columna clave v谩lida, analizar duplicados por esa columna
    # -------------------------------------------------
    # if clave analiza que clave no sea None
    # and (y)
    if clave and clave in df.columns:
    # clave in df.columns-- >que clave sea una columna existente dentro de las columnas del dataframe
    # si no le paso ninguna columna no va a querer encontrar duplicados por columna
    # y si me equivoco y le paso una columna que no existe en el dataframe, tampoco ingresara al if.
        # Contar cu谩ntas filas tienen valores repetidos en esa columna
        dup_key = df[clave].duplicated(keep=False).sum()
        print(f"Duplicados por clave '{clave}':", dup_key)

        # Si existen duplicados, mostrar cu谩les son los valores m谩s repetidos
        if dup_key > 0:
            # Filtrar filas donde esa clave est茅 duplicada
            # df[clave].duplicated(keep=False) devuelve True donde el valor se repite
            duplicados_ordenados = (
                df[df[clave].duplicated(keep=False)][clave]
                .value_counts()                # Cuenta cu谩ntas veces aparece cada valor
                .sort_values(ascending=False)   # Ordena de mayor a menor (m谩s duplicados arriba)
            )

            print("\n Top valores duplicados m谩s frecuentes:")
            # Mostrar solo los primeros 10 (los m谩s repetidos)
            display(duplicados_ordenados.head(40))
        else:
            print(f"No se encontraron duplicados en la clave '{clave}'.")
    else:
        # Si la clave no fue pasada o no existe en el DataFrame
        if clave:
            print(f"La clave '{clave}' no existe en el DataFrame.")
        else:
            print("No se indic贸 una clave para analizar duplicados por columna.")
#fin de def calidad
```

##Reporte comparativo de cambios

```python

# ============================================
# Se define una funci贸n que lee los tres DataFrames limpios (o los recibe en memoria)
# y muestra un resumen comparativo de nulos, duplicados y tipos de datos.
# ============================================

def reporte_calidad_global(dfs, nombres):
    """
    Crea un resumen de calidad de varios DataFrames.

    Par谩metros:
      dfs: lista de DataFrames (por ejemplo [ventas_clean, clientes_clean, marketing_clean])
      nombres: lista de nombres correspondientes (["VENTAS", "CLIENTES", "MARKETING"])
    """
    resumen = []
    #zip-->es una funci贸n incorporada de Python que une elementos de dos (o m谩s) iterables
    # por ejemplo, listas, tuplas o cualquier objeto iterable en pares ordenados.
    for df, nombre in zip(dfs, nombres):                 # con zip recorro dos iterables con un solo FOR
        nulos = df.isna().sum().sum()                    # Total de valores nulos,  por columnas y total, por eso el doble sum
        duplicados = df.duplicated(keep=False).sum()     # Total de filas duplicadas
        columnas = len(df.columns)                       # Cantidad de columnas
        filas = len(df)                                  # Cantidad de registros

        resumen.append({"Dataset": nombre,"Filas": filas,"Columnas": columnas,"Nulos totales": nulos,"Duplicados": duplicados,})

    reporte = pd.DataFrame(resumen)
    #display(reporte)
    return reporte

```

##ETAPA 2:2. Transformaci贸n de datos (alto rendimiento)**


### Objetivo: construir una tabla de rendimiento por producto y quedarnos s贸lo con los productos de alto rendimiento.

Conceptos clave:

<h3>Transformaci贸n de datos:</h3> son operaciones que crean/derivan nuevas columnas (por ejemplo ingreso = precio * cantidad), normalizan formatos (texto/fechas/n煤meros) o filtran filas seg煤n un criterio.

<h3>M茅trica de ingreso:</h3> para ventas, una m茅trica t铆pica es ingreso por registro = precio * cantidad. Luego podemos agregar por producto (sumar ingresos y unidades) para medir rendimiento total por producto.

<h3>Agregaci贸n:</h3> es resumir muchas filas en pocas, aplicando funciones como sum(), mean(), count() agrupando por una clave (ej., producto).
Ej.: ingreso total por producto = suma de todos los ingresos de ese producto.

<h3>Percentil:</h3> el percentil 80 (P80) es un valor tal que el 80% de los datos est谩n por debajo o igual a ese valor y el 20% restante por encima.

Si ingreso_total P80 = 120.000, significa que el 80% de los productos tienen ingreso_total  120.000 y el 20%  120.000.

<h3>Alto rendimiento:</h3> aqu铆 lo definimos como top 20% de productos seg煤n ingreso_total (>= P80). Es un criterio com煤n cuando no hay umbrales de negocio expl铆citos.
Alternativas v谩lidas: top-K (p. ej. top 50 productos), percentil 75 (P75) o un umbral fijo de negocio (p. ej., >= $100.000/mes), o score estandarizado (z-score).

<h3>Plan paso a paso:</h3>

Detectar la columna de producto (tolerando distintos nombres: producto, id_producto, sku, articulo).

Calcular ingreso por registro = precio * cantidad.

Agregar por producto para obtener m茅tricas (ingreso_total, unidades, precio_promedio, registros).

Calcular P80 con quantile(q=0.80).

Filtrar productos con ingreso_total >= P80.

Ordenar de mayor a menor.
#-------------------------------------------------------------------------------

```python
# Objetivo:
# - Detectar los productos con mejor desempe帽o econ贸mico (top 20% por ingreso total).
# - Aplicar transformaci贸n: calcular ingreso, agregar por producto y filtrar.

# FUNCION para detectar una columna de interes.
def encontrar_columna(df, candidatos):
    """
    Busca la primera columna cuyo nombre contenga alguno de los patrones dados.
    - df: DataFrame de pandas.
    - candidatos: lista de patrones (min煤sculas).
    """
    # Recorremos todas las columnas del DataFrame
    for c in df.columns:

        # Convertimos el nombre de la columna a min煤sculas
        # Esto se hace para comparar sin importar si est谩 escrito con may煤sculas o min煤sculas
        nombre = c.lower()

        # Verificamos si alguna palabra (patr贸n) de la lista 'candidatos'
        # est谩 contenida dentro del nombre de la columna
        # 'any()' devuelve True apenas encuentra una coincidencia
        if any(p in nombre for p in candidatos):
            # Si encontramos una coincidencia, devolvemos el nombre original de la columna
            return c

    # Si termina el bucle y no se encontr贸 ninguna coincidencia, devolvemos None
    return None
```

```python


# 1锔 Detectar la columna de producto con la funcion anterior.
prod_col = encontrar_columna(df_ventas_clean, ["producto", "id_producto", "sku", "articulo", "art铆culo"])
if prod_col is None:
    raise ValueError("No se encontr贸 columna de producto. Renombr谩 una columna a 'producto' o similar.")
print(prod_col)

```

```python
# 2锔 Calcular ingreso por registro = precio * cantidad
#     () conjunto, es para escribir en varias filas
ventas_perf = (
    df_ventas_clean
    .assign(
        ingreso = df_ventas_clean["precio"] * df_ventas_clean["cantidad"]
        # assign(**nuevas_col): crea nuevas columnas y devuelve una copia del DF.
        # Alternativa: ventas_clean["ingreso"] = ventas_clean["precio"] * ventas_clean["cantidad"]
    )
)
#esta linea comentada es igual que la linea multiple de arriba
#ventas_perf = df_ventas_clean.assign(ingreso = ventas_clean["precio"] * ventas_clean["cantidad"])
#esta otra linea agrega a ventas_clean una columna nueva ingreso y le asigna precio*cantidad
#ventas_clean["ingreso"] = ventas_clean["precio"] * ventas_clean["cantidad"]     ESTA ES LA MAS ENTENDIBLE!!!!!!!!
```

```python

# 3锔 Agregar m茅tricas por producto creando un nuevo dataframe "resumen-prod".
# CLASE 8, 1:30'////1:36'////
#recordar prod_col es el nuevo df con la clumna "Ingreso".
# ventas_perf es el nuevo df donde inserte la columna "Ingreso".
#quiero que ese df sea agrupado por columnas, en este caso por una, prod_col
resumen_prod = (
    ventas_perf
    # 1) Agrupamos el DataFrame por una o varias columnas clave
    .groupby(
        by=prod_col,    # Columna (str) o lista de columnas (list[str]) que define los grupos.
        dropna=False,   # False  NO descarta filas donde la clave de grupo tenga NaN; crea un grupo para NaN.
        as_index=False, # False  las columnas de agrupaci贸n quedan como columnas normales (no pasan al 铆ndice).
        observed=False  # Solo aplica si 'prod_col' es Categorical:
                        #   False  incluye categor铆as NO observadas (posibles pero sin filas);
                        #   True   solo categor铆as que aparecen en los datos (m谩s r谩pido y compacto).
    )
    # 2) Agregamos (resumimos) columnas num茅ricas por cada grupo
    .agg(
        ingreso_total=('ingreso', 'sum'),   # Suma de 'ingreso' por grupo (skipna=True por defecto).
        unidades=('cantidad', 'sum'),       # Suma de 'cantidad' por grupo.
        precio_promedio=('precio', 'mean'), # Promedio simple de 'precio' por grupo (ignora NaN).
        registros=('ingreso', 'size')       # N煤mero de filas en el grupo (cuenta TODO, incluso NaN).
    )
)

#*************************************************************************************************************************************
#se puede escribir asi:
#resumen_prod = ventas_perf.groupby(by=prod_col).agg(ingreso_total=('ingreso', 'sum'), unidades=('cantidad', 'sum'), precio_promedio=('precio', 'mean'), registros=('ingreso', 'size'))
#*************************************************************************************************************************************
```

```python


# 4锔 Calcular percentil 80 de ingreso_total.

# OJO VUELVO A df "resumen_prod" y filtro la columna "ingreso_total" y busco el
# q80. -------------------------------------------------------- La funci贸n
# quantile() nos permite obtener el valor de un percentil. En este caso,
# queremos saber el ingreso que separa al 80% de los productos con menores
# ingresos del 20% con mayores ingresos.

p80_ingreso = resumen_prod["ingreso_total"].quantile(
    q=0.80,                # q indica el percentil deseado (0.80 = 80% de los datos por debajo)
    interpolation="linear" # si el percentil no coincide exactamente con un valor real del dataset,
                           # 'linear' interpola entre los dos valores vecinos.
                           # Ejemplo: si el 80% cae entre 4000 y 5000,
                           # calcula un valor proporcional, por ejemplo 4200.
                           # Otros m茅todos posibles:
                           #  - 'lower': toma el menor de los dos valores (4000)
                           #  - 'higher': toma el mayor (5000)
                           #  - 'nearest': el m谩s cercano al percentil
                           #  - 'midpoint': el punto medio exacto (4500)
)

# En resumen:
# - quantile calcula el valor l铆mite de un percentil.
# - q define qu茅 percentil queremos.
# - interpolation define c贸mo se calcula cuando el valor no est谩 exactamente en los datos.
# El resultado (p80_ingreso) es el ingreso total que marca el l铆mite superior del 80% de los productos.

print(f"P80 de ingreso_total: {float(p80_ingreso):,.2f}")
```

```python
# 5锔 Filtrar los productos "de alto rendimiento" y ordenarlos
# -------------------------------------------------------------------
# Contexto: `resumen_prod` es un DataFrame con m茅tricas por producto,
# y `p80_ingreso` es el percentil 80 de la columna "ingreso_total".
# Objetivo: quedarnos con los productos cuyo ingreso_total est谩 en el 20% superior
# (ingreso_total >= p80_ingreso) y luego ordenarlos de mayor a menor por ingreso y unidades.
# en una sola fila
#ventas_top = resumen_prod.query("ingreso_total >= @p80_ingreso",engine="python").sort_values(by=["ingreso_total", "unidades"], ascending=[False, False], na_position="last", ignore_index=True
ventas_top = (
    resumen_prod
    # ---------------------------------------------------------------
    # .query(expr, *, inplace=False, engine='python'|'numexpr').
    # .query  PERMITE FILTRAR DATOS DE UNA TABLA.
    #   - Aplica un filtro usando una expresi贸n estilo SQL-simple.
    #   - `expr` es un string que se eval煤a sobre los nombres de las columnas.
    #   - Para usar variables de Python (no columnas), se antepone '@' (ej.: @p80_ingreso).
    #   - NaN en comparaciones (>, >=, ==, etc.) se comportan como False  esas filas no pasan el filtro.
    #   - engine='python': interpreta la expresi贸n con Python puro (compatible siempre).
    #   - engine='numexpr': si est谩 instalado, acelera operaciones num茅ricas vectorizadas.
    #   - inplace: False (por defecto) devuelve un DF nuevo; True modifica el DF original (menos recomendado en cadenas).
    .query(
        "ingreso_total >= @p80_ingreso",  # expr: filtra filas donde ingreso_total es al menos el umbral del P80
        engine="python"                   # motor de evaluaci贸n (usar 'numexpr' si lo ten茅s y quer茅s performance)
        # Notas de sintaxis de `expr`:
        #    Operadores l贸gicos: and / or / not   (tambi茅n valen &, |, ~ con par茅ntesis).
        #    Strings deben ir entre comillas: canal == 'Online'
        #    Columnas con espacios o caracteres raros: usar `backticks`, ej.: `nombre producto` == 'X'
        #    Ejemplos:
        #       "ingreso_total >= @p80_ingreso and unidades >= 10"
        #       "`nombre producto`.str.contains('Promo')"
        #       "precio_promedio.between(1000, 3000, inclusive='both')"
    )
    # ---------------------------------------------------------------
    # .sort_values(by, axis=0, ascending=True|[...], inplace=False,
    #              kind='quicksort'|'mergesort'|'heapsort'|'stable',
    #              na_position='last'|'first', ignore_index=False, key=None)
    #   - Ordena por una o varias columnas.
    #   - `by`: str o lista de str con las columnas a ordenar.
    #   - `ascending`: bool o lista de bool (una por cada columna en `by`).
    #   - `na_position`: d贸nde ubicar NaN ('last' o 'first').
    #   - `ignore_index`: si True, reasigna el 铆ndice 0..n-1 en el resultado.
    #   - `kind`: algoritmo de ordenamiento (mergesort es estable si necesit谩s preservar empates).
    #   - `key`: funci贸n que transforma los valores antes de ordenar (p. ej., key=lambda s: s.str.normalize(...)).
    .sort_values(
        by=["ingreso_total", "unidades"],  # primero ordena por ingreso_total, luego desempata por unidades
        ascending=[False, False],          # ambos en orden descendente (mayor  menor)
        na_position="last",                # coloca NaN al final (煤til si alguna m茅trica qued贸 en NaN)
        ignore_index=True                  # reindexa el resultado secuencialmente (0..n-1)
        # Variantes 煤tiles:
        #    ascending=True                 # orden ascendente
        #    ascending=[False, True]        # primero desc, luego asc para el segundo criterio
        #    kind='mergesort'               # orden estable (respeta el orden de aparici贸n en empates)
        #    key=lambda s: s.str.lower()    # ordenar texto sin distinci贸n de may煤sculas/min煤sculas
    )
)
print(ventas_top.head(10))
# Resultado:
# `ventas_top` contiene solo los productos cuyo ingreso_total >= p80_ingreso,
# ordenados de mayor a menor por ingreso_total y, ante empates, por unidades.
```

##ETAPA 2: 3. Agregaci贸n: Resumir las ventas por *CATEGORIA* de producto y analizar los ingresos generados.

### Conceptos y plan

Objetivo: construir un resumen por categor铆a de producto con m茅tricas 煤tiles (ingreso total, unidades, cantidad de ventas, ticket promedio).

Conceptos clave:

Agregaci贸n: operaci贸n que resume muchas filas en menos filas, aplicando funciones (sum, mean, count, etc.) despu茅s de agrupar por una clave (aqu铆, la categor铆a).

Categor铆a de producto: atributo que agrupa productos similares (ej., Electr贸nica, Hogar).

Ticket promedio por venta: ingreso_total / ventas (d贸nde ventas es el conteo de filas en esa categor铆a). Indica el importe medio facturado por cada transacci贸n/registro en la categor铆a.

Nota: esto no es el precio promedio del producto; ese ya se calcula con mean sobre precio.

Consideraciones: outliers pueden distorsionar promedios; a veces conviene mirar tambi茅n la mediana (median).


```python
# ============================================
# 8) AGREGACIN: resumen por categor铆a
#ventas : dataset original
#ventas_clean: dataset original despues de limpiar datos
#ventas_perf: dataset de ventas_clean con el campo ingreso que es precio*cantidad de cada factura
#resumen_prod: dataset de ventas_perf agrupado por producto y sumado el ingreso de cada venta = ingreso_total
#ventas_top: el detaset de resumen_prod, que tiene productos agrupados, pero solo me quedo con los productos mas vendidos (top del 20%)
#ventas_cat: el dataset de ventas_perf que se copio y se llama ventas_cat
#resumen_cat: del dataset de ventas_perf agrupo por categoria de producto y sumo ingreso_total

ventas_cat = ventas_perf.copy() # ventas limpias+columna ingreso (ventas_clean)

# 3) Agregaci贸n por categor铆a con groupby + agg
resumen_cat = (
    ventas_cat
    .groupby(
        by="categoria",      # Puede ser string o lista de strings si quisi茅ramos agrupar por varias columnas.
        dropna=False,    # Mantener grupo NaN (si hay filas sin categor铆a).
        as_index=False   # Dejar la categor铆a como columna normal (y no como 铆ndice).
        # observed: si cat_col es 'category' y queremos mostrar solo categor铆as presentes  True.
    )
    .agg(
        ingreso_total=('ingreso', 'sum'),   # Suma total de ingresos por categor铆a.
        unidades=('cantidad', 'sum'),       # Unidades totales vendidas en la categor铆a.
        ventas=('ingreso', 'size'),         # Cantidad de registros/filas (ventas) en la categor铆a.
        precio_promedio=('precio', 'mean')  # Precio promedio observado en la categor铆a.
        # Otras funciones 煤tiles: 'median','max','min','std','var','nunique'...
    )
    .sort_values(
        by='ingreso_total', # Ordenar por ingreso total
        ascending=False,    # Descendente: mayores arriba
        na_position='last', # NaN al final
        ignore_index=True   # Reindexar desde 0
    )
)

# 4) Ticket promedio por venta = ingreso_total / ventas
resumen_cat = resumen_cat.assign(
    ticket_promedio_por_venta = resumen_cat['ingreso_total'] / resumen_cat['ventas']
    # assign: crea/reescribe columnas. Alternativa: resumen_cat['ticket_promedio_por_venta'] = ...
)


print("Resumen por categor铆a (ordenado por ingreso_total):")
display(resumen_cat.head(20))
```

## Rango intercuart铆lico (IQR) del Ingreso Totales por Categoria.
El IQR mide la dispersi贸n del 50% central de los datos y se calcula como:

[ IQR = Q3 - Q1 ]

Es menos sensible a los outliers.

 Interpretaci贸n:
El IQR refleja el rango donde se encuentran las ventas normales, sin verse afectado por los outliers.

```python
# df sigue siendo *resumen_cat* es decir *ingreso_total* es de c/categoria.
q1 = df['ingreso_total'].quantile(0.25)
q2 = df['ingreso_total'].quantile(0.5)
q3 = df['ingreso_total'].quantile(0.75)
iqr = q3 - q1

print("Q1:", q1)
print("Mediana (Q2):", q2)
print("Q3:", q3)
print("IQR:", iqr)
```

## Detecci贸n de Outliers de Ingreso Total de c/categoria usando el IQR
Regla general:

L铆mite inferior = Q1  (1.5  IQR)

L铆mite superior = Q3 + (1.5  IQR)

```python
lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr

outliers = df[(df['ingreso_total'] < lim_inf) | (df['ingreso_total'] > lim_sup)]

print("L铆mite inferior:", lim_inf)
print("L铆mite superior:", lim_sup)
print("\nOutliers detectados:")
print(outliers)
```
